# Processing columns with mixed variables

The aim of this part is to extarct numerical and categorical parts of a variable then store them in two different columns.


```python
# Note that I just extract the first number when we have multiple numbers for a family
def number_extractor(cl):
    if(pd.isnull(cl)):
        return np.nan
    a=str(cl)
    a=a[1:]
    l=[]
    k=0
    for i in a:
        if(i.isdigit()==False):
            k=1
        if(i.isdigit()==True and k==0):
            l.append(i)
    return ''.join(l)

data['Cabin_numerical']=data['Cabin'].apply(number_extractor)

submission['Cabin_numerical']=submission['Cabin'].apply(number_extractor)
```

```python
data['Cabin_numerical']=pd.to_numeric(data['Cabin_numerical'])
submission['Cabin_numerical']=pd.to_numeric(submission['Cabin_numerical'])
```

## Random sampling technique
### Creating additional variable
I want to impute NA random sampling technique in Age and ticket and use additional variable to indicate missingness

```python
def impute_na_random(df, X_train, column_name):
    #df1 could be one of the Trrain, Test and submission
    df1=df.copy()
    # we have to extact specific amount of number from non-elements of X_train 
    a=X_train[column_name].dropna().sample(df1[column_name].isnull().sum(), random_state=0)
    a.index=df1[df1[column_name].isnull()].index
    df1.loc[df1[column_name].isnull(), column_name]=a
    return df1[column_name]
```

```python
#add variable to indicate missingness
for i in [X_test, X_train, submission]:
    for j in ['Age', 'Ticket_numerical']:
        i[j+'NaN']=np.where(i[j].isnull(),1,0)

# Replace Nan by random sampling        
for i in [X_test, X_train, submission]:
    for j in ['Age', 'Ticket_numerical']:
        i[j]=impute_na_random(i, X_train, j)
```
